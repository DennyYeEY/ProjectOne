# Analysis of My Use Case

---

## Use Case

- My use case is to find the top exporters of the US based on the shipping company, consignee company, and foreign country.

---

### Silver Layer

- Based on the data provided, I decided to use the 2019-2020 header.csv, consignee.csv, and shipper.csv.
- Through normalization, I broke the header table into a vessel, secondary, cargo_measurement, and port table.

---

### Gold Layer

- After I finished cleaning and modifying my silver data, I decided began to take off unnecessary columns and tables that were unrelated to my use case.
- In addition, I began to de-normalize my tables into star schema.
