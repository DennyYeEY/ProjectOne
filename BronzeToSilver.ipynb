```python
import pandas as pd
pd.set_option('display.max_columns', None)

from datetime import datetime, UTC
import os

import glob
```

```python
def buildTables():

    def read_and_concat(file_pattern):
        files = glob.glob(file_pattern)
        dfs = [
            pd.read_csv(
                f,
                engine="python"     
            )
            for f in files
        ]
        return pd.concat(dfs, ignore_index=True)

    # ----- HEADER -----
    headerDf2019 = read_and_concat("BronzeLayerData/header/2019/*.csv")
    headerDf2019.to_csv("SilverLayerData/header2019.csv", index=False)

    headerDf2020 = read_and_concat("BronzeLayerData/header/2020/*.csv")
    headerDf2020.to_csv("SilverLayerData/header2020.csv", index=False)

    # ----- CONSIGNEE -----
    consigneeDf = read_and_concat("BronzeLayerData/consignee/*.csv")
    consigneeDf.to_csv("SilverLayerData/consignee.csv", index=False)

    # ----- SHIPPER -----
    shipperDf = read_and_concat("BronzeLayerData/shipper/*.csv")
    shipperDf.to_csv("SilverLayerData/shipper.csv", index=False)
```

## Transformation_Log Function

```python
import csv

LOG_FILE = "transformation_log.csv"

# Function to log transformations
def log_transformation(
    input_source: str, # File or table name that is read
    output_source: str, # File or table name that is written to
    action_description: str, # Description of the transformation performed
    performed_by: str, # Name performing the transformation
    log_path: str = LOG_FILE
):
    file_exists = os.path.exists(log_path)

    with open(log_path, mode="a", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)

        if not file_exists:
            writer.writerow([
                "timestamp",
                "input_file_or_table",
                "output_file_or_table",
                "action_description",
                "performed_by"
            ])

        writer.writerow([
            datetime.now(UTC).isoformat(),
            input_source,
            output_source,
            action_description,
            performed_by
        ])
```

# Cleaning up header

### Converting the weight into Kilograms, then dropping weight_unit

```python
def convert_weights(df, output_path):
    WEIGHT_CONVERSIONS = {
        "Metric Ton": 1000,
        "Pounds": 0.453592,
        "Long Ton": 1016.05,
        "Short Ton": 907.185
    }

    mask = df['weight_unit'].isin(WEIGHT_CONVERSIONS)

    # Looks up the conversion factor for each weight unit and applies it to the weight column, then rounds and converts to int
    df.loc[mask, "weight"] = (
        df.loc[mask, "weight"] * df.loc[mask, "weight_unit"].map(WEIGHT_CONVERSIONS)
    ).round().astype(int)

    # Deletes the weight_unit column as it's no longer needed after conversion
    drop_columns(df, output_path, ['weight_unit'])

    # Save the transformed DataFrame to the specified output path
    df.to_csv(output_path, index=False)

    return
```

### Mode_of_transportation

- [Vessel, containerized] -> C
- [Vessel, non-container] -> N

```python
def replaceTransportation(df, output_path):
    df = df.copy() 

    # Withinin the "mode_of_transportation" column, we replace "Vessel, containerized" with "C" and "Vessel, non-container" with "N". We also strip any leading/trailing whitespace and ensure the column is of string type.
    df["mode_of_transportation"] = (
        df["mode_of_transportation"]
        .astype(str)
        .str.strip()
        .replace({
            "Vessel, containerized": "C",
            "Vessel, non-container": "N",
        })
    )
    df.to_csv(output_path, index=False)
    return
```

### secondaryNotifyParty

```python
def createSecondaryNotifyParty(df, output_path):
    df = df.copy()

    # Look through all the columm names and keep only the ones that start with "secondary_notify_party_"
    notify_cols = [
        col for col in df.columns
        if col.startswith("secondary_notify_party_")
    ]

    # Melt the DataFrame to have one row per notify party
    melted = df.melt(
        id_vars=["identifier"],
        value_vars=notify_cols,
        var_name="notify_party_column",
        value_name="notify_party_name"
    )

    # Drops rows where notify_party_name is null
    melted = melted.dropna(subset=["notify_party_name"])

    # Extract the sequence number from the notify_party_column
    melted["notify_party_sequence"] = (
        melted["notify_party_column"]
        .str.extract(r"_(\d+)$")
        .astype(int)
    )
    
    result = melted[
        ["identifier", "notify_party_sequence", "notify_party_name"]
    ]

    result.to_csv(output_path, index=False)

    return result
```

### Port Table

```python
def createPort(df, output_path, port_columns):
    df = df.copy()

    # Concatenate all port columns into a single Series, ignoring any missing columns
    all_ports = pd.concat(
        [df[col] for col in port_columns if col in df.columns],
        ignore_index=True
    )

    # Converts to string
    all_ports = (
        all_ports.astype("string")
        .str.strip()
    )

    # Drops null or empty port names
    all_ports = all_ports[all_ports.notna() & (all_ports != "")]

    # Get unique port names
    unique_ports = pd.Series(all_ports.unique(), dtype="string")

    # Deterministic IDs: sort so port_id assignment is stable across runs
    unique_ports = unique_ports.sort_values(kind="stable").reset_index(drop=True)

    # Create a DataFrame with port_id and port_name columns
    port_df = pd.DataFrame({
        "port_id": range(1, len(unique_ports) + 1),
        "port_name": unique_ports
    })

    port_df.to_csv(output_path, index=False)

    return port_df
```

```python
# Create port table with port_id and port_name columns for all unique ports across the relevant port columns in the header table
createPort(headerDf, 'SilverLayerData/port.csv', ['port_of_unlading', 'foreign_port_of_lading_qualifier', 'foreign_port_of_lading', 'port_of_destination', 'foreign_port_of_destination_qualifier', 'foreign_port_of_destination'])
log_transformation('SilverLayerData/header.csv', 'SilverLayerData/port.csv', 'Created port table with port_id and port_name columns', 'Denny Ye')
```

```python
def addPortIdsToHeader(df, port_df, mapping, output_path):
    df = df.copy()

    # Build a lookup dict: port_name -> port_id
    port_lookup = dict(zip(port_df["port_name"], port_df["port_id"]))

    # For each source column in the mapping, creating a new column with port_id
    for source_col, id_col in mapping.items():
        if source_col not in df.columns:
            continue

        cleaned = (
            df[source_col]
            .astype("string")
            .str.strip()
        )

        df[id_col] = cleaned.map(port_lookup)
    
    df.to_csv(output_path, index=False)

    return
```

```python
# Mapping for columns
mapping = {
    'port_of_unlading': 'port_of_unlading_id',
    'foreign_port_of_lading_qualifier': 'foreign_port_of_lading_qualifier_id',
    'foreign_port_of_lading': 'foreign_port_of_lading_id',
    'port_of_destination': 'port_of_destination_id',
    'foreign_port_of_destination_qualifier': 'foreign_port_of_destination_qualifier_id',
    'foreign_port_of_destination': 'foreign_port_of_destination_id'
}

# Add port_id columns to header table based on the port names and port_id in the port table
addPortIdsToHeader(headerDf, portDf, mapping, 'SilverLayerData/header.csv')
log_transformation('SilverLayerData/header.csv', 'SilverLayerData/header.csv', 'Added port_id columns to header table based on the port names and port_id in the port table', 'Denny Ye')
```

### Vessel Table

```python
def createVessel(df, output_path):
    df = df.copy()

    # Select vessel_country_code and vessel_name columns
    vessel_df = df[
        ["vessel_country_code", "vessel_name"]
    ].copy()

    # Clean up the columns
    vessel_df["vessel_country_code"] = (
        vessel_df["vessel_country_code"]
        .astype("string")
        .str.strip()
    )

    vessel_df["vessel_name"] = (
        vessel_df["vessel_name"]
        .astype("string")
        .str.strip()
    )

    # Drops rows where vessel_name is null
    vessel_df = vessel_df[
        vessel_df["vessel_name"].notna()
        & (vessel_df["vessel_name"] != "")
    ]

    # Get unique vessel_contry_code and vessel_name combinations
    vessel_df = vessel_df.drop_duplicates()

    # Deterministic ordering so vessel_id is stable across runs
    vessel_df = (
        vessel_df
        .sort_values(
            by=["vessel_country_code", "vessel_name"],
            kind="stable"
        )
        .reset_index(drop=True)
    )

    # Create vessel_id starting from 1
    vessel_df.insert(
        0,
        "vessel_id",
        range(1, len(vessel_df) + 1)
    )

    vessel_df.to_csv(output_path, index=False)

    return 
```

```python
# Creates vessel table from Header csv
createVessel(headerDf, 'SilverLayerData/vessel.csv')
log_transformation('SilverLayerData/header.csv', 'SilverLayerData/vessel.csv', 'Created vessel table with vessel_id, vessel_country_code, vessel_name columns', 'Denny Ye')
```

```python
def addVesselIdToHeader(df, vessel_df, output_path):
    df = df.copy()

    # Normalize join columns in header
    df["vessel_country_code"] = (
        df["vessel_country_code"]
        .astype("string")
        .str.strip()
    )

    df["vessel_name"] = (
        df["vessel_name"]
        .astype("string")
        .str.strip()
    )

    vessel_df = vessel_df.copy()

    # Normalize join columns in vessel_df
    vessel_df["vessel_country_code"] = (
        vessel_df["vessel_country_code"]
        .astype("string")
        .str.strip()
    )

    vessel_df["vessel_name"] = (
        vessel_df["vessel_name"]
        .astype("string")
        .str.strip()
    )

    # Build lookup using merge on vessel_country_code and vessel_name to get vessel_id into header table
    df = df.merge(
        vessel_df[
            ["vessel_id", "vessel_country_code", "vessel_name"]
        ],
        on=["vessel_country_code", "vessel_name"],
        how="left"
    )

    df.to_csv(output_path, index=False)

    return

```

```python
# Add vessel_id to header table
addVesselIdToHeader(headerDf, vesselDf, 'SilverLayerData/header.csv')
log_transformation('SilverLayerData/header.csv', 'SilverLayerData/header.csv', 'Added vessel_id to header table by merging with vessel table on vessel_country_code and vessel_name', 'Denny Ye')
```

# Shipper Table

Need to deduplicate shipper data

```python
def deduplicate(df, output_path, subset_cols):
    df = df.copy()

    df = df.drop_duplicates(subset=subset_cols)

    df.to_csv(output_path, index=False)

    return
```

Create a temporary staging for shipper.csv and header.csv

```python
def createStaging(header_df, data_df):
    
    # Does a left merge from header_df to shipper_df on identifier to bring the shipper columns into the staging table
    staging = header_df[["identifier"]].merge(
        data_df,
        on="identifier",
        how="left",
        validate="one_to_one"
    )
    return staging
```

Building the shipper table

```python
def buildShipperFromStaging(staging_df, output_path):
    shipper_cols = [
        "shipper_party_name",
        "shipper_party_address_1",
        "shipper_party_address_2",
        "shipper_party_address_3"
    ]

    # Keep only the rows where shipper_party_name is not null
    shipper_df = (
        staging_df[shipper_cols]
        .dropna(subset=["shipper_party_name"])
        .drop_duplicates()
        .reset_index(drop=True)
    )

    # Deterministic IDs
    shipper_df = shipper_df.sort_values(shipper_cols, kind="stable")
    shipper_df.insert(0, "shipper_id", range(1, len(shipper_df) + 1))

    shipper_df.to_csv(output_path, index=False)
    return
```

Create shipper lookup table

```python
def createShipperLookup(staging_df, shipper_df):
    shipper_cols = [
        "shipper_party_name",
        "shipper_party_address_1",
        "shipper_party_address_2",
        "shipper_party_address_3"
    ]

    lookup = staging_df.merge(
        shipper_df,
        on=shipper_cols,
        how="left",
        validate="many_to_one"
    )

    
    # Ensure shipper_id is a nullable integer, not float
    lookup["shipper_id"] = lookup["shipper_id"].astype("Int64")


    return lookup[["identifier", "shipper_id"]]
```

Add shipper_id to header

```python
def addShipperIdToHeader(header_df, shipper_lookup, output_path):
    newHeader = header_df.merge(
        shipper_lookup,
        on="identifier",
        how="left",
        validate="one_to_one"
    )

    newHeader.to_csv(output_path, index=False)
    return
```
